# 1 WORDLE / WORD CLOUD

- First we make a text file with president JFK's speech. 
- Then we use the WordCloud to make the wordcloud.

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)


with open("jfk_speech_moon_landing.txt", "r") as file:
    text = file.read()

# Create a set of stopwords to exclude from the word cloud
stopwords = set(STOPWORDS)

# Generate the word cloud
wordcloud = WordCloud(width=1600, height=800, background_color='white', stopwords=stopwords, font_step=4).generate(text)

# Plot the word cloud
plt.figure(figsize=(16, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
# plt.tight_layout(pad=0)

# Show the plot
plt.show()

```
